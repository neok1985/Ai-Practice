```python
#시리즈의 개별 원소에 함수 매핑 : 시리즈.apply(매핑함수)
import pandas as pd
import seaborn as sns

titanic=sns.load_dataset('titanic')
```

```python
#titanic 에서 'age'와 'fare' 컬럼만 추출해서 df생성
df=titanic.loc[:,['age','fare']]
df.head()
```

```python
#매핑할 함수 정의 : 10을 더해서 return 하는 함수 정의
def add_10(n):
    return n + 10
#두 개의 값을 받아 두 수의 합을 return 하는 함수 정의
def add_2obj(a,b):
    return a+b
print(add_10(20))#10+20의 값을 출력
print(add_2obj(10,20))#10+20의 값을 출력
```

```python
#df의 'age' 컬럼에 10을 더한 결과를 sr1에 저장 후 데이터확인
sr1=df['age'].apply(add_10)

sr2=df['age'].apply(add_2obj,b=10)
sr2.head()
#lambda 함수로 정의
sr3=df['age'].apply(lambda x: add_10(x))
# sr3.head()
#lambda 함수를 바로 정의해서 사용
sr4=df['age'].apply(lambda x:x+10)
sr4.head()
```

```python
#데이터프레임의 각 원소에 함수 매핑: 데이터프레임.applymap(매핑함수)
df_map=df.applymap(add_10)
df_map.head()
```

```python
#데이터프레임 각 원소에 함수 매핑 :데이터프레임.apply(매핑함수,axis=0or1)
df_1=df.apply(add_10,axis=0)


df_2=df.apply(lambda x: x['age']+x['fare'],axis=1)
# print(df_1.head(),df_2.head())
df_2.iloc[885:890,]
# print(df_1{['age','fare']}+df_2{['age','fare']})
```

```python
#데이터프레임 객체에 함수매핑 : 데이터프레임.pipe(매핑함수)
#각 열의 Nan 자료를 찾아 -> df 반환
def missing_value(x):
    return x.isnull()
#각 열의 Nan 갯수 반환:데이터프레임->시리즈 반환
def missing_count(x):
    return missing_value(x).sum()
#데이터 프레임 Nan의 총 갯수 : 값을 반환
def total_number_missing(x):
    return missing_count(x).sum()
```

```python
df_1=df.pipe(missing_value)
print(df_1.head())
sr_1=df.pipe(missing_count)
print(sr_1)
value=df.pipe(total_number_missing)
value
```

```python
value1=df.apply(total_number_missing)
value1
```

```python
#2단계:그룹연산을 그룹의 객체 적용
df=titanic.loc[:,['age','class','fare','survived']]
grouped=df.groupby(['class'])#1단계

#표준편차를 그룹별로 구함
std_all=grouped.std() #2단계
print(std_all)

#'fare'컬럼에 대해서만 표준편차 구함
grouped['fare'].std()
```

```python
#사용자 정의 함수를 그룹 객체에 적용 : 그룹객체.agg(매핑함수)
agg_max_min=grouped[['age','fare']].agg(lambda x: max(x)-min(x))
agg_max_min
```

```python
#동시에 여러 개의 함수를 적용: 그룹객체.agg([함수1,함수2,...])
#컬럼마다 다른 함수를 적용: 그룹객체.agg({'열':함수,'열2':함수2,...})
agg_all=grouped.agg(['min','max','mean'])
agg_all
```

```python
agg_def=grouped.agg({'age':'mean','fare':['max','std']})
agg_def
```
