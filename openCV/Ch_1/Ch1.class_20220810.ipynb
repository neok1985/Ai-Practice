```python
import numpy as np
import cv2
import sys
import matplotlib.pyplot as plt
cv2.__version__
```

```python
img=cv2.imread('./fig/puppy.bmp',cv2.IMREAD)
# img_resize=cv2.resize(img,(640*2,480*2),interpolation=cv2.INTER_LINEAR)
# cv2.imwrite('img_puppy2resize.png',img_resize)
if img is None:
    print('image read failed')
    sys.exit()
cv2.namedWindow('image'cv2.WINDOW_NORMAL)
cv2.imshow('image',img)
# cv2.imshow('image_resize',img_resize)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img3=cv2.imread('./fig/puppy.bmp',cv2.IMREAD_GRAYSCALE)

if img3 is None:
    print('image read failed')
    sys.exit()
cv2.namedWindow('image_3t_hwindow',cv2.WINDOW_NORMAL)
cv2.imshow('image_3t_hwindow',img3)
while True:
    key=cv2.waitKey()
    if key == 27 or key == ord('q') :
        break
cv2.destroyAllWindows()
```

```python
img=cv2.imread('./fig/puppy.bmp',cv2.IMREAD_COLOR)
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
if img is None:
    print('imge read failed')
    sys.exit()
plt.figure(figsize=(15,20))    
plt.subplot(131),plt.imshow(img),plt.axis('off')
plt.subplot(132),plt.imshow(img_rgb),plt.axis('off')
plt.subplot(133),plt.imshow(img_gray,cmap='gray'),plt.axis('off')
# plt.axis('off')
plt.show()

# cv2.imshow('image',img)
# cv2.imshow('image2',img_gray)

# while True:
#     key1=cv2.waitKey()
#     if key1 ==27 or key == ord('q'):
#         break
# cv2.destroyAllWindows()
```
